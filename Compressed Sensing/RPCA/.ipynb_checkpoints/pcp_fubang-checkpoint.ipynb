```python
import math
import numpy as np
from PIL import Image
import sys
import glob
import matplotlib.pyplot as plt
import time
```

```python
def l1_norm(X):
    """
    L1-Norm of X
    """
    return np.linalg.norm(np.reshape(X, -1), ord=1)

def frobenius_norm(X):
    """
    Frobenius norm of X
    """
    return np.linalg.norm(X, ord='fro')

def check_conv(M, L, S):
    """
    Return the value of ||M-L-S||_F / ||M||_F
    """
    error = frobenius_norm(M - L - S) / frobenius_norm(M)
    return error

def s(X, tau):
    """
    The shrinkage operator
    X: M -L + 1/mu * Y in the case of Paper
    tau: lamda / mu
    Return: the new S for the next iteration
    """
    V = np.copy(X).reshape(X.size)
    for i in range(V.size):
        V[i] = math.copysign(max(abs(V[i]) - tau, 0), V[i])
    return V.reshape(X.shape)

def d(X, tau):
    """
    The singular value thresholding operator.
    X: M -L + 1/mu * Y
    tau: 1/mu
    Return: the new L for the next iteration
    """
    U, eps, V = np.linalg.svd(X, full_matrices=False)
    return np.dot(U, np.dot(np.diag(s(eps, tau)), V))
```

```python
def rpca_ALM(M, error_threshold):
    """ 
    Principal component pursuit by augmented Lagrange multiplier algorithm.
    M: the matrix of raw data
    error_threshold: the threshold of error => ||M-L-S||_F / ||M||_F <= delta
    L: the low rank component, matrix, true data
    S: the sparse component, matrix 
    Errs: the error list of each iteration
    """
    M_shape = M.shape
    L = np.zeros(M_shape)
    S = np.zeros(M_shape)
    Y = np.zeros(M_shape)
    Errs = []
    mu = (M_shape[0] * M_shape[1]) / (4.0 * l1_norm(M))
    lamb = max(M_shape) ** -0.5
    while True:
        L = d(M - S + (mu**-1) * Y, mu ** (-1))
        S = s(M - L + (mu**-1) * Y, lamb / mu)
        Y = Y + mu * (M - L - S)
        err = check_conv(M, L, S)
        Errs.append(err)
        if err <= error_threshold:
            break
    return L, S, Errs
```

## 1. Exact recovery from varying fractions of error

```python
def data_generate(n, sup_s):
    '''
    Generate the simulation data where L is 0.05*n rank, and S is sup_s*n^2 sparse.
    n: M is square and its dimension is (n*n)
    sup_s: The support of S is sup_s*n^2
    Return the simulation of L and S
    '''
    r = int(0.05 * n)
    k = int(sup_s * n ** 2)
    X = np.random.normal(0, 1, (n, r))
    Y = np.random.normal(0, 1, (n, r))
    L_0 = X.dot(Y.T)
    a = np.arange(n * n)
    np.random.shuffle(a)
    S_0 = np.zeros(n * n)
    for i in a[:k]:
        S_0[i] = 2 * np.random.choice(2, 1, p=[0.5, 0.5])[0] - 1
    S_0 = S_0.reshape((n, n))
    return L_0, S_0
```

```python
def recovery(d_ns, sup_s):
    '''
    Recover the data from the simulation L_0+S_0
    d_ns: the list of the dimension of M
    sup_s: The support of S is sup_s*n^2
    '''
    for d_n in d_ns:
        L_0, S_0 = data_generate(d_n, sup_s)
        print('The rank of the L_0 is: %r' %np.linalg.matrix_rank(L_0))
        print('The support of S_0 is: %r' %S_0.nonzero()[0].shape[0])
        L_1, S_1, Errs_1 = rpca_ALM(L_0 + S_0, 1e-5)
        print('The rank of the L_pred is: %r' %np.linalg.matrix_rank(L_1))
        print('The support of S_pred is: %r' %S_1.nonzero()[0].shape[0])
        print('The number of iteration: %d' %len(Errs_1))
        print('The final error: %r \n' %Errs_1[-1])
```

```python
recovery([500, 1000, 2000, 3000], 0.05)
```

```python
recovery([500, 1000, 2000, 3000], 0.1)
```

## 2. Background modeling from surveillance video

```python
def bitmap_to_mat(bitmap_seq):
    '''
    bitmap_seq: the filepath of bitmap data
    Return the matrix of monochrome image
    '''
    matrix = []
    shape = None
    for bitmap_file in bitmap_seq:
        img = Image.open(bitmap_file).convert("L")
        if shape is None:
            shape = img.size
        assert img.size == shape
        img = np.array(img.getdata())
        matrix.append(img)
    return np.array(matrix), shape[::-1]

def do_plot(ax, img, shape):
    '''
    Plot the monochrome image from the matrix.
    img: the matrix of a bitmap data
    '''
    ax.cla()
    ax.imshow(img.reshape(shape), cmap="gray", interpolation="nearest")
```

### Load the data

```python
M, shape = bitmap_to_mat(glob.glob("data/*.bmp")) # the filepath of raw data
```

```python
print('The shape of raw data Matrix M is %s' % (M.shape,))
print('The shape of each image is: %s' %(shape,))
```

```python
t = time.time()
L, S, Errs = rpca_ALM(M, 1e-5) #change it to modify the threshold of error
print("Iteration costs:%.2f" %(time.time()-t))
```

```python
plt.plot(Errs)
plt.title('The convergence of PCP Algorithm')
plt.xlabel('The number of iteration')
plt.ylabel('Error')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 8))
fig.subplots_adjust(left=0, right=1, hspace=0, wspace=0.01)
n_fig = 50 # change it to check the effect of different image
do_plot(axes[0], M[n_fig], shape)
axes[0].set_title("raw")
do_plot(axes[1], L[n_fig], shape)
axes[1].set_title("low rank")
do_plot(axes[2], np.abs(S[n_fig]), shape)
axes[2].set_title("sparse")
plt.show()
```

From the images above, we can see that the raw data is splited into two components: 
1. The low-rank component (the escalator);
2. The sparse component (the passerbys).

The result is similar with that in the paper, the threshold of error that we set is $10^{-5}$.

Within 200 rounds of iteration, we can already get a pretty good result.

```python
print('The support of S is:%r' %np.abs(S).nonzero()[0].shape)
print('The support of M is:%r' %np.abs(M).nonzero()[0].shape)
print('The support of L is:%r' %np.abs(L).nonzero()[0].shape)
```

```python
print('The rank of L is: %r' %np.linalg.matrix_rank(L))
print('The rank of M is: %r' %np.linalg.matrix_rank(M))
print('The rank of S is: %r' %np.linalg.matrix_rank(S))
```

As the result shown above, $L$ is low-rank and not sparse, while $S$ is sparse and not low-rank, which satisfies the requirement of the question.
